````
#!/usr/bin/env python3
import base64
import requests
import pandas as pd

url = 'https://api.github.com/repos/mikeizbicki/datasets/contents/csv/uci/breast-cancer-wisconsin.data'
req = requests.get(url)

if req.status_code == requests.codes.ok:
    req = req.json()  # the response is a JSON
    # req is now a dict with keys: name, encoding, url, size ...
    # and content. But it is encoded with base64.
    content = base64.standard_b64decode(req['content'])
else:
    print('Content was not found.')

with open('breast-cancer-wisconsin.data', 'w') as f:
    f.write(content.decode('ascii'))

dataset = pd.read_table('breast-cancer-wisconsin.data', delimiter=',', header=None)
print(dataset.head(5))

!rm -rf breast-cancer-wisconsin.data
!ls -la
````
